```python
import numpy as np
from PIL import Image, ImageDraw
import random
import matplotlib.pyplot as plt
from operator import attrgetter

class PolygonIndividual:
    def __init__(self, polygons, target_array):
        self.target_array = target_array
        if polygons is None:
            self.polygons = self.random_polygons()
        else:
            self.polygons = polygons
        self.fitness = self.get_fitness()

    def random_polygons(self):
        num_polygons = 50  # Adjust the number of polygons
        polygons = []
        for _ in range(num_polygons):
            polygon = {
                'vertices': [(random.randint(0, self.target_array.shape[1]), random.randint(0, self.target_array.shape[0])) for _ in range(3)],  # Triangle
                'color': tuple(np.random.randint(0, 256, 3))  # Random color
            }
            polygons.append(polygon)
        return polygons

    def draw_image(self):
        image = Image.new('RGB', (self.target_array.shape[1], self.target_array.shape[0]), (255, 255, 255))
        draw = ImageDraw.Draw(image, 'RGBA')
        for polygon in self.polygons:
            draw.polygon(polygon['vertices'], fill=polygon['color'])
        return image

    def get_fitness(self):
        generated_image = np.array(self.draw_image())
        return np.sum((generated_image - self.target_array) ** 2)

    def mutate(self, mutation_rate):
        for polygon in self.polygons:
            if random.random() < mutation_rate:
                polygon['vertices'] = [(random.randint(0, self.target_array.shape[1]), random.randint(0, self.target_array.shape[0])) for _ in range(3)]
            if random.random() < mutation_rate:
                polygon['color'] = tuple(np.random.randint(0, 256, 3))
        self.fitness = self.get_fitness()

    @staticmethod
    def crossover(parent1, parent2):
        child_polygons = []
        for p1, p2 in zip(parent1.polygons, parent2.polygons):
            if random.random() < 0.5:
                child_polygons.append(p1)
            else:
                child_polygons.append(p2)
        return PolygonIndividual(polygons=child_polygons, target_array=parent1.target_array)


class PolygonPopulation:
    def __init__(self, size, target_array):
        self.size = size
        self.target_array = target_array
        self.individuals = [PolygonIndividual(target_array=target_array) for _ in range(size)]

    def evolve(self, gens, xo_prob, mut_prob, select, xo, mutate, elitism):
        for gen in range(gens):
            new_pop = sorted(self.individuals, key=attrgetter('fitness'))
            if elitism:
                new_pop = new_pop[:1]
            else:
                new_pop = []

            while len(new_pop) < self.size:
                parent1, parent2 = select(self), select(self)
                if random.random() < xo_prob:
                    offspring = xo(parent1, parent2)
                else:
                    offspring = random.choice([parent1, parent2])

                if random.random() < mut_prob:
                    offspring.mutate(mut_prob)

                new_pop.append(offspring)

            self.individuals = new_pop
            best_individual = min(self.individuals, key=attrgetter('fitness'))
            print(f"Best fitness of gen #{gen + 1}: {best_individual.fitness}")

    def get_best_individual(self):
        return min(self.individuals, key=attrgetter('fitness'))


def tournament_sel(population, tour_size=3):
    tournament = [random.choice(population.individuals) for _ in range(tour_size)]
    return min(tournament, key=attrgetter('fitness'))


target_image = Image.open('target_image.jpg').convert('RGB')
target_array = np.array(target_image)

population_size = 50
generations = 1000
xo_prob = 0.7
mut_prob = 0.1
elitism = True

population = PolygonPopulation(size=population_size, target_array=target_array)

population.evolve(
    gens=generations,
    xo_prob=xo_prob,
    mut_prob=mut_prob,
    select=tournament_sel,
    xo=PolygonIndividual.crossover,
    mutate=PolygonIndividual.mutate,
    elitism=elitism
)

best_individual = population.get_best_individual()
best_image = best_individual.draw_image()

best_image.save('best_image.jpeg')
best_image.show()

plt.imshow(best_image)
plt.axis('off')
plt.show()
```
